```python
import os
import pdfplumber
```

```python
pdf = False
for _pdf in os.listdir('Docs'):
    if "Datacenter" in _pdf:
        pdf = pdfplumber.open(f'Docs/{_pdf}')
        break
```

```python
l_class = list()
d_pages = dict()
reqs = list(map(lambda x: x.strip(), open("requirment_list.txt", "r", encoding="utf-8").readlines()))

for req in reqs:
    if "*" in req:
        c = ' '.join(req.split()[1:-1])
        start_page, end_page = req.split()[-1].split(",")
        l_class.append(c)
        d_pages[c] = (int(start_page) - 1, int(end_page))
```

```python
for c in l_class:
    print(c, d_pages[c][0], d_pages[c][1])
    for page in pdf.pages[d_pages[c][0]:d_pages[c][1]]:
        tables = page.extract_tables()
        for x in tables:
            for y in x:
                _y = y[0]                   
                if _y and ("-" in _y) and (_y.split("-")[-1].isdigit()):
                    sub = "-".join(_y.split("-")[:-1]).strip()

    break
```
